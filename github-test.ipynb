```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Sample 100 data points
diamonds_sample = diamonds.sample(n=100, random_state=42)

# Create a scatter plot with carat on the x-axis and price on the y-axis, colored by cut
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diamonds_sample, x='carat', y='price', hue='cut', palette='viridis')

# Add labels and title
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Diamonds: Carat vs Price by Cut')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Sample 100 data points
diamonds_sample = diamonds.sample(n=1000, random_state=42)

# Create a FacetGrid with cut as the facet variable
g = sns.FacetGrid(diamonds_sample,
                  col='cut',
                  col_wrap=5,
                  height=4,
                  hue='cut', palette='viridis')

# Map the scatter plot onto the grid
g.map(sns.scatterplot, 'carat', 'price')

# Add labels and title
g.set_axis_labels('Carat', 'Price')
g.fig.suptitle('Diamonds: Carat vs Price Faceted by Cut', y=1.05)

# Show plot
plt.show()
```
